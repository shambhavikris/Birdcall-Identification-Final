```python
import pandas as pd
import numpy as np
from numpy import asarray
from numpy import savez_compressed
import os
import tensorflow as tf
import pydub
```

```python
dir = '/content/drive/My Drive/lala1/train_audio/DataUpload2'
output_dir = '/content/drive/My Drive/lala1/output2-npz-full'
ms_of_data = 30000*30

for d in os.listdir(dir):#class

  folder = os.path.join(dir, d)

  flag = 0

  for f in os.listdir(folder):#on file in class
    # each .mp3 file
    fn = os.path.join(folder, f)
    print('fn', fn)
    if not fn.endswith('mp3'):
      continue
    x = pydub.AudioSegment.from_file(file=fn)
    #os.remove(fn)
    x = x.set_frame_rate(30000)
    x = x.set_sample_width(1)
    x = np.array(x.get_array_of_samples(), dtype=np.float32)
    x = np.trim_zeros(x)
    num = len(x)//ms_of_data
    if num==0:
      x = np.pad(x, (0, ms_of_data - len(x)))
    x = x[0:(len(x)//ms_of_data)*ms_of_data]
    x = np.array(np.split(x, len(x)//ms_of_data))
    num2 = 1

    for i in x:
      i = np.expand_dims(i, axis=-1)
      if flag == 0:
          flag = 1 #first done
          subset_total = i

      else:
          subset_total = np.concatenate((subset_total, i), axis=1)
  
  filename = output_dir + '/' + d + '.npz'
  print('saving from ', d, 'class to', filename)
  savez_compressed(filename, subset_total)
  print('shape of this ^ is ', subset_total.shape)
```

```python
print('done')
```
